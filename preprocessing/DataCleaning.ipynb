```python
import pandas as pd
import re 
import string
#from spellchecker import SpellChecker
from tqdm import tqdm

df = pd.read_csv("/content/drive/MyDrive/ML_proj/train.csv")
df_test = pd.read_csv("/content/drive/MyDrive/ML_proj/test.csv")
```

# Remove URLs


```python
def remove_URL(text):
    pattern = r'https?://\S+|www\.\S+'
    replaced_text = re.sub(pattern, '', text)
    return replaced_text


df['text']=df['text'].apply(lambda x : remove_URL(x))
```

# Removing emoticons, symbols, pictographs, transport, map symbols



```python
def remove_emojis(text):
    emojis = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols, pictographs
                               u"\U0001F680-\U0001F6FF"  # transport, map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags
                               "]+", flags=re.UNICODE)
    
    cleaned_text = emojis.sub(r'', text)
    
    return cleaned_text


df['text']=df['text'].apply(lambda x : remove_emojis(x))
```

```python
def remove_punct(text):
    punctuation = string.punctuation
    
    for pu in punctuation:
        text = text.replace(pu, '')
    
    return text

df['text']=df['text'].apply(lambda x : remove_punct(x))
```

# Spell checking

```python
spell = SpellChecker()

def correct_spelling(text):
    misspelled_words = spell.unknown(text.split())
    corrected_text = []
    for word in tqdm(text.split()):
        if word in misspelled_words:
            correction = spell.correction(word)
            if correction is not None:
                corrected_text.append(correction)
            else:
                corrected_text.append(word)
        else:
            corrected_text.append(word)
    return " ".join(corrected_text)
        
df['text']=df['text'].apply(lambda x : correct_spelling(x))
```

# Save the cleaned data

```python
df.to_csv(path_or_buf = '/content/drive/MyDrive/ML_proj/clean_data.csv')
df_test.to_csv(path_or_buf = '/content/drive/MyDrive/ML_proj/clean_testset.csv')
```
