```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image


# load data
df = pd.read_csv('/content/drive/MyDrive/LT_SHARED_FOLDER/train.csv')
```

```python
# Concatenate all the text data into a single string
text = ' '.join(df['text'].astype(str))
```

```python
# create a mask with the Twitter logo
twitter_mask = 255-np.array(Image.open("/content/81609.png"))

# create the WordCloud object
wc = WordCloud(background_color="black", max_words=2000, mask=twitter_mask, contour_width=3, contour_color='lightblue')

# generate the word cloud
wc.generate(text)

# display the word cloud
plt.figure(figsize=(13, 13))

plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
